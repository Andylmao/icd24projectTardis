```python
import cv2
import os
import numpy as np

# Ruta de tu carpeta en Google Drive
carpeta_imagenes = '/content/drive/MyDrive/DataScience/TardisImages/tardis/heterotardigrada'

# Configuración de CLAHE
def aplicar_clahe(img_path):
    # Leer imagen en escala de grises
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    # Crear objeto CLAHE
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
    # Aplicar CLAHE
    return clahe.apply(img)

# Crear carpeta para guardar imágenes procesadas
carpeta_procesadas = '/content/drive/MyDrive/DataScience/TardisImages/tardisAugment/heterotaraugm'
if not os.path.exists(carpeta_procesadas):
    os.makedirs(carpeta_procesadas)
```

```python
for nombre_imagen in os.listdir(carpeta_imagenes):
    if nombre_imagen.endswith(('.png', '.jpg', '.jpeg')):
        ruta_imagen = os.path.join(carpeta_imagenes, nombre_imagen)
        imagen_procesada = aplicar_clahe(ruta_imagen)

        ruta_guardado = os.path.join(carpeta_procesadas, nombre_imagen)
        cv2.imwrite(ruta_guardado, imagen_procesada)
        print(f'Imagen procesada y guardada en {ruta_guardado}')
```

```python
from PIL import Image
import os

# Ruta de la carpeta de imágenes originales y la de subimágenes
carpeta_original = '/content/drive/MyDrive/DataScience/TardisImages/tardis/heterotardigrada'
carpeta_subimagenes = '/content/drive/MyDrive/DataScience/TardisImages/tardis320x320/hetero320x320'
os.makedirs(carpeta_subimagenes, exist_ok=True)

def dividir_y_redimensionar(imagen_path, nuevo_tamano=(320, 320)):
    img = Image.open(imagen_path)
    ancho, alto = img.size

    # Dividir la imagen en 4 partes 800x600 cada una
    subimagenes = [
        img.crop((0, 0, ancho // 2, alto // 2)),
        img.crop((ancho // 2, 0, ancho, alto // 2)),
        img.crop((0, alto // 2, ancho // 2, alto)),
        img.crop((ancho // 2, alto // 2, ancho, alto))
    ]

    # Redimensionar y guardar cada subimagen
    for i, subimg in enumerate(subimagenes):
        subimg = subimg.resize(nuevo_tamano, Image.LANCZOS)
        nombre_subimagen = f"{os.path.splitext(os.path.basename(imagen_path))[0]}_parte{i+1}.png"
        ruta_guardado = os.path.join(carpeta_subimagenes, nombre_subimagen)
        subimg.save(ruta_guardado)
        print(f'Subimagen guardada en {ruta_guardado}')

# Aplicar a todas las imágenes en la carpeta
for nombre_imagen in os.listdir(carpeta_original):
    if nombre_imagen.endswith(('.png', '.jpg', '.jpeg')):
        ruta_imagen = os.path.join(carpeta_original, nombre_imagen)
        dividir_y_redimensionar(ruta_imagen)
```

```python
import cv2
import os
import numpy as np

# Directorios de entrada y salida
input_dir = '/content/drive/MyDrive/DataScience/TardisImages/tardis/heterotardigrada'  # Carpeta con las imágenes originales
output_dir = '/content/drive/MyDrive/DataScience/TardisImages/tardAugmRotBri/heteAugmrot'  # Carpeta para las imágenes aumentadas
os.makedirs(output_dir, exist_ok=True)  # Crea la carpeta de salida si no existe

# Parámetros de aumento
rotation_angles = [45 * i for i in range(8)]  # Rotaciones en múltiplos de 22.5 grados (16 imágenes por rotación)
brightness_factors = [.6, 1.4]  # Factores para ajustar el brillo: 0.8 (más oscuro), 1.0 (sin cambio), 1.2 (más brillante)

# Función para cambiar el brillo de una imagen
def adjust_brightness(image, factor):
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    hsv = np.array(hsv, dtype=np.float64)
    hsv[:, :, 2] = hsv[:, :, 2] * factor  # Escala el canal de brillo (V) por el factor
    hsv[:, :, 2][hsv[:, :, 2] > 255] = 255  # Asegúrate de que el valor esté entre 0 y 255
    hsv = np.array(hsv, dtype=np.uint8)
    return cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)

# Procesa cada imagen en la carpeta de entrada
for filename in os.listdir(input_dir):
    img_path = os.path.join(input_dir, filename)
    image = cv2.imread(img_path)

    if image is not None:
        # Aplica rotación y cambio de brillo
        for angle in rotation_angles:
            # Rotación
            (h, w) = image.shape[:2]
            center = (w // 2, h // 2)
            rotation_matrix = cv2.getRotationMatrix2D(center, angle, 1.0)
            rotated_image = cv2.warpAffine(image, rotation_matrix, (w, h))

            for factor in brightness_factors:
                # Ajuste de brillo
                bright_image = adjust_brightness(rotated_image, factor)

                # Guarda la imagen con rotación y brillo ajustado
                output_filename = f"{os.path.splitext(filename)[0]}_rot{int(angle)}_bright{factor}.png"
                output_path = os.path.join(output_dir, output_filename)
                cv2.imwrite(output_path, bright_image)
                print(f"Imagen guardada: {output_path}")
    else:
        print(f"No se pudo cargar la imagen: {filename}")
```
